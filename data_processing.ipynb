```python
import pandas as pd
```

```python
essayAI = pd.read_csv('ai_manual_data.csv')
essayHuman = pd.read_csv('human_data_general.csv')
df_Hu = essayHuman.copy()
df_AI = essayAI.copy()
```

```python
df_Hu['Text'] = df_Hu['text']
df_Hu.drop(columns= ['text', 'title', 'label', 'source_license'], inplace= True)
df_Hu.head()
```

```python
df_Hu['Author'] = 'Human'
df_Hu.head()
```

```python
df_AI.drop_duplicates(keep= False, inplace= True)
df_AI['Text'] = df_AI['text']
df_AI.drop(columns= ['text', 'label', 'source'], inplace= True)
df_AI['Author'] = 'AI'
df_AI.head()
```

```python
df = pd.concat([df_AI, df_Hu], ignore_index= True)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```
