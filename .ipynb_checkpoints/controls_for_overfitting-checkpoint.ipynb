```python
import pandas as pd
```

```python
data = pd.read_csv('data_for_preprocessing.csv')
df = data.copy()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
label_mapping = {'Human': 0, 'AI': 1}
df['label_num'] = df['Author'].map(label_mapping)
df.head()
```

```python
df.drop(columns = ['Author', 'Unnamed: 0'], inplace = True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.Text
y = df.label_num
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model_pipeline = make_pipeline(
    TfidfVectorizer(stop_words='english', max_features=5000), 
    LogisticRegression()
)
```

```python
model_pipeline.fit(X_train, y_train)
```

```python
y_pred = model_pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Doğruluk Oranı (Accuracy): %{accuracy * 100:.2f}")
print("\nDetaylı Rapor:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['İnsan (0)', 'AI (1)'], yticklabels=['İnsan (0)', 'AI (1)'])
plt.xlabel('Tahmin Edilen')
plt.ylabel('Gerçek Olan')
plt.title('Confusion Matrix (Hata Matrisi)')
plt.show()
```

```python
print("\n--- DETAYLI SINIFLANDIRMA RAPORU ---")
print(classification_report(y_test, y_pred, target_names=['İnsan', 'AI']))
```

```python
from sklearn.model_selection import cross_val_score
print("\n--- ÇAPRAZ DOĞRULAMA (5 KATLI) ---")
# Veriyi 5'e bölüp her parçada test eder
scores = cross_val_score(model_pipeline, X, y, cv=5)
print(f"5 Farklı Testin Ortalaması: %{scores.mean()*100:.2f}")
print(f"Skorlar: {scores}")
```

```python
y_prob = model_pipeline.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate (Yanlış Alarm)')
plt.ylabel('True Positive Rate (Doğru Tespit)')
plt.title('ROC Eğrisi')
plt.legend(loc="lower right")
plt.show()
```
